## Below Adds Tools to Unity Catalog

### Get customer information based on name

```python
%sql

CREATE OR REPLACE FUNCTION get_customer_info_by_name(
    first_name_input STRING,
    last_name_input STRING COMMENT 'customer first and last name used to retrieve customer information'
)
RETURNS TABLE (
    customer_id BIGINT,
    first_name STRING,
    last_name STRING,
    city STRING,
    state STRING,
    tenure_years DOUBLE,
    churn_risk_score BIGINT,
    customer_value_score BIGINT
)
COMMENT 'Returns the customer record matching the provided first and last name. Includes its ID, firstname, lastname and more.'
RETURN (
    SELECT customer_id, first_name, last_name, city, state, tenure_years, churn_risk_score, customer_value_score 
    FROM customers
    WHERE first_name = first_name_input 
    AND last_name = last_name_input
);
```

```python
%sql SELECT * FROM get_customer_info_by_name('Danielle', 'Johnson');
```

### Get customer information based on id

```python
%sql
    
CREATE OR REPLACE FUNCTION get_customer_info_by_id(
    customer_id_input BIGINT COMMENT 'customer first and last name used to retrieve customer information'
)
RETURNS TABLE (
    customer_id BIGINT,
    first_name STRING,
    last_name STRING,
    city STRING,
    state STRING,
    tenure_years DOUBLE,
    churn_risk_score BIGINT,
    customer_value_score BIGINT
)
COMMENT 'Returns the customer record matching the provided first and last name. Includes its ID, firstname, lastname and more.'
RETURN (
    SELECT customer_id, first_name, last_name, city, state, tenure_years, churn_risk_score, customer_value_score 
    FROM customers
    WHERE customer_id = customer_id_input
);
```

```python
%sql SELECT * FROM get_customer_info_by_id(10001);
```
